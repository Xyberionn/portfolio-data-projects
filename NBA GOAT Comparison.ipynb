```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np
plt.style.use('dark_background')
```

```python
players_names = ["Michael Jordan", "LeBron James", "Kobe Bryant","Magic Johnson","Larry Bird"]
```

```python
def get_player_id(name):
    player_list = players.get_players()
    for player in player_list:
        if player['full_name'] == name:
            return player['id']
    return None
```

```python
all_data = []

for player_name in players_names:
    player_id = get_player_id(player_name)
    
    if player_id:
        stats = playercareerstats.PlayerCareerStats(player_id=player_id)
        career_data = stats.get_data_frames()[0]
        print(f"{player_name}:")
        display(career_data)

        career_data['PLAYER_NAME'] = player_name

        all_data.append(career_data)

df_goat_stats = pd.concat(all_data, ignore_index=True)
df_goat_stats.columns
```

```python

stats = ['PTS', 'AST', 'DREB', 'OREB', 'TOV', 'STL', 'BLK']

for stat in stats:
    df_goat_stats[f'{stat}_per_game'] = df_goat_stats[stat] / df_goat_stats['GP']
df_goat_stats
```

```python

per_game_stats = [f'{stat}_per_game' for stat in stats] + ['FG_PCT', 'FG3_PCT', 'FT_PCT']

player_avg = df_goat_stats.groupby("PLAYER_NAME")[per_game_stats].mean().reset_index()
player_avg
```

```python
# PTS_per_game, FG_PCT, FT_PCT : 0.6, 0.2, 0.2

def normalize(series, reverse=False):
    norm = 100 * (series - series.min()) / (series.max() - series.min())
    return 100 - norm if reverse else norm

scoring_df = player_avg[['PLAYER_NAME', 'PTS_per_game', 'FG_PCT', 'FT_PCT']].copy()

scoring_df['PTS_norm'] = normalize(scoring_df['PTS_per_game'])
scoring_df['FG_norm'] = normalize(scoring_df['FG_PCT'])
scoring_df['FT_norm'] = normalize(scoring_df['FT_PCT'])

scoring_df['SCORING_SCORE'] = (
    0.6 * scoring_df['PTS_norm'] +
    0.2 * scoring_df['FG_norm'] +
    0.2 * scoring_df['FT_norm']
)
scoring_result = scoring_df[['PLAYER_NAME', 'SCORING_SCORE']].sort_values(by='SCORING_SCORE', ascending=False)
print(scoring_result)
```

```python
playmaking_df = player_avg[['PLAYER_NAME', 'AST_per_game', 'TOV_per_game']].copy()

playmaking_df['AST_norm'] = normalize(playmaking_df['AST_per_game'])

playmaking_df['TOV_norm'] = normalize(playmaking_df['TOV_per_game'], reverse=True)

playmaking_df['PLAYMAKING_SCORE'] = (
    0.7 * playmaking_df['AST_norm'] +
    0.3 * playmaking_df['TOV_norm']
)
playmaking_result = playmaking_df[['PLAYER_NAME', 'PLAYMAKING_SCORE']].sort_values(by='PLAYMAKING_SCORE', ascending=False)
print(playmaking_result)
```

```python
defense_df = player_avg[['PLAYER_NAME', 'STL_per_game', 'BLK_per_game', 'DREB_per_game']].copy()

defense_df['STL_norm'] = normalize(defense_df['STL_per_game'])
defense_df['BLK_norm'] = normalize(defense_df['BLK_per_game'])
defense_df['DREB_norm'] = normalize(defense_df['DREB_per_game'])

defense_df['DEFENSE_SCORE'] = (
    0.4 * defense_df['STL_norm'] +
    0.4 * defense_df['BLK_norm'] +
    0.2 * defense_df['DREB_norm']
)
defense_result = defense_df[['PLAYER_NAME', 'DEFENSE_SCORE']].sort_values(by='DEFENSE_SCORE', ascending=False)
print(defense_result)
```

```python
shooting_df = player_avg[['PLAYER_NAME', 'FG_PCT', 'FG3_PCT', 'FT_PCT']].copy()

shooting_df['FG_norm'] = normalize(shooting_df['FG_PCT'])
shooting_df['FG3_norm'] = normalize(shooting_df['FG3_PCT'])
shooting_df['FT_norm'] = normalize(shooting_df['FT_PCT'])

shooting_df['SHOOTING_SCORE'] = (
    0.4 * shooting_df['FG_norm'] +
    0.4 * shooting_df['FG3_norm'] +
    0.2 * shooting_df['FT_norm']
)
shooting_result = shooting_df[['PLAYER_NAME', 'SHOOTING_SCORE']].sort_values(by='SHOOTING_SCORE', ascending=False)
print(shooting_result)
```

```python
categories = ['SCORING_SCORE', 'PLAYMAKING_SCORE', 'DEFENSE_SCORE', 'SHOOTING_SCORE']
radar_df = scoring_df[['PLAYER_NAME', 'SCORING_SCORE']].merge(
    playmaking_df[['PLAYER_NAME', 'PLAYMAKING_SCORE']], on='PLAYER_NAME'
).merge(
    defense_df[['PLAYER_NAME', 'DEFENSE_SCORE']], on='PLAYER_NAME'
).merge(
    shooting_df[['PLAYER_NAME', 'SHOOTING_SCORE']], on='PLAYER_NAME'
)

players = radar_df['PLAYER_NAME'].tolist()

N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1] 

fig, ax = plt.subplots(figsize=(9, 8), subplot_kw=dict(polar=True))

color_mapping = {
    'Michael Jordan': '#E74C3C',  # Red
    'LeBron James': '#8E44AD',    # Purple
    'Kobe Bryant': '#F1C40F',     # Gold
    'Larry Bird': '#27AE60',      # Green
    'Magic Johnson': '#FFFFFF'    # White
}

for player in players:
    values = radar_df.loc[radar_df['PLAYER_NAME'] == player, categories].values.flatten().tolist()
    values += values[:1]  
   
    color = color_mapping.get(player, '#000000') 

    ax.plot(angles, values, label=player, color=color)
    ax.fill(angles, values, alpha=0.1, color=color)

ax.set_theta_offset(np.pi / 2) 
ax.set_theta_direction(-1) 

radius = 125 
for i, category in enumerate(categories):
    ax.text(angles[i], radius, category, horizontalalignment='center', size=9, color='white')

ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))

max_seasons_player = df_goat_stats.groupby(df_goat_stats.index)['SEASON_ID'].count().idxmax()
max_seasons = df_goat_stats[df_goat_stats.index == max_seasons_player]['SEASON_ID'].nunique()

color_mapping = {
    'Michael Jordan': '#E74C3C',  # Red
    'LeBron James': '#8E44AD',    # Purple
    'Kobe Bryant': '#F1C40F',     # Gold
    'Larry Bird': '#27AE60',      # Green
    'Magic Johnson': '#FFFFFF'    # White
}

for idx, player_name in enumerate(players_names):
    player_data = df_goat_stats[df_goat_stats.index == player_name]
    
    player_data['PTS_PER_GAME'] = player_data['PTS'] / player_data['GP']
    
    player_data['SEASON_ORDER'] = range(1, len(player_data) + 1)

    color = color_mapping.get(player_name, '#7F8C8D')

    plt.plot(player_data['SEASON_ORDER'], player_data['PTS_PER_GAME'], label=player_name, color=color, marker='o')

season_labels = [f"{i}." for i in range(1, max_seasons + 1)]
plt.xticks(ticks=range(1, max_seasons + 1), labels=season_labels)

plt.title("Per Points by Season", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("PTS per Game", fontsize=12)
plt.legend(title="Players", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
